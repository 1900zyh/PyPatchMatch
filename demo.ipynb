```python
from patchmatch_inpaint import PyPatchMatch_inpaint
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
```

```python
input = io.imread('image_files/forest/forest.bmp')
mask = io.imread('image_files/forest/forest_pruned.bmp')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
plt.imshow(input)
plt.subplot(122)
plt.imshow(mask)
```

```python
%%time
output = PyPatchMatch_inpaint(input, mask[:,:,0]/255, 1)

plt.imshow(output)
```
